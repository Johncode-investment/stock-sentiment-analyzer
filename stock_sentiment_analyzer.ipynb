```python
import yfinance as yf
from textblob import TextBlob
import matplotlib.pyplot as plt

# Choose your stock ticker
stock_ticker = "AAPL"  # Apple Inc.

# Fetch recent stock data
data = yf.download(stock_ticker, period="7d", interval="1d")
print("\nRecent Stock Data:\n")
print(data[['Open', 'Close']])

# Simulated headlines
headlines = [
    "Apple launches new iPhone 15 with breakthrough camera features",
    "Investors worry about Apple's declining iPad sales",
    "Apple stock surges after strong quarterly results",
    "Apple faces antitrust lawsuit over App Store policies",
    "Mixed reactions to Apple's new MacBook design"
]

# Analyze sentiment
print("\nSentiment Analysis:\n")
sentiment_scores = []
for headline in headlines:
    blob = TextBlob(headline)
    score = blob.sentiment.polarity
    sentiment_scores.append(score)
    print(f"{headline} --> Sentiment Score: {score:.2f}")

# Plot results
plt.figure(figsize=(10,5))
plt.plot(sentiment_scores, marker='o', color='orange', label='Headline Sentiment')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Sentiment Scores for Apple News Headlines")
plt.xlabel("Headline Index")
plt.ylabel("Sentiment Polarity (-1 to +1)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
